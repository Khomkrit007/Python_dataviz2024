##2.สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน

H0 (สมมติฐานหลัก): ค่าเฉลี่ย Chlo_b ของแต่ละ station ไม่แตกต่างกันระหว่างเดือนต่างๆ



Ha (สมมติฐานทางเลือก): ค่าเฉลี่ย Chlo_b ของแต่ละ station แตกต่างกันระหว่างเดือนต่างๆ อย่างน้อยหนึ่งคู่

เนื่องจากเราต้องการเปรียบเทียบค่าเฉลี่ยของกลุ่มข้อมูลมากกว่าสองกลุ่ม (เดือนต่างๆ) เราสามารถใช้ การวิเคราะห์ความแปรปรวนทางเดียว (One-way ANOVA)

```python
# ลบเครื่องหมาย "_" ออกจากคอลัมน์ month
folk_df['month'] = folk_df['month'].str.rstrip('_')

# แสดง DataFrame เพื่อตรวจสอบผลลัพธ์
print(folk_df)
```

```python
# สร้างตารางค่าเฉลี่ย Chlo_b โดยจัดกลุ่มตาม station และ month
chlo_b_mean_table = folk_df.groupby(['station', 'month'])['Chlo_b'].mean().unstack()

# แสดงตาราง
print(chlo_b_mean_table)
```

```python
from scipy import stats

# ดึงข้อมูล Chlo_b สำหรับแต่ละเดือนจากตาราง chlo_b_mean_table
apr_data = chlo_b_mean_table['Apr'].dropna()
aug_data = chlo_b_mean_table['Aug'].dropna()
dec_data = chlo_b_mean_table['Dec'].dropna()
jan_data = chlo_b_mean_table['Jan'].dropna()
jul_data = chlo_b_mean_table['July'].dropna()
jun_data = chlo_b_mean_table['June'].dropna()
may_data = chlo_b_mean_table['May'].dropna()
nov_data = chlo_b_mean_table['Nov'].dropna()
oct_data = chlo_b_mean_table['Oct'].dropna()
sep_data = chlo_b_mean_table['Sep'].dropna()

# ทำการทดสอบ One-way ANOVA
f_statistic, p_value = stats.f_oneway(apr_data, aug_data, dec_data, jan_data, jul_data, jun_data, may_data, nov_data, oct_data, sep_data)

print("F-statistic:", f_statistic)
print("P-value:", p_value)
```

p-value น้อยกว่าระดับนัยสำคัญทั่วไป (เช่น 0.05) อย่างมาก เราจึงปฏิเสธสมมติฐานหลัก (H0) และยอมรับสมมติฐานทางเลือก (Ha)

มีหลักฐานเพียงพอที่จะสรุปว่าค่าเฉลี่ย Chlo_b ของแต่ละ station แตกต่างกันระหว่างเดือนต่างๆ อย่างน้อยหนึ่งคู่
