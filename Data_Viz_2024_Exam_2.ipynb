```python
path_to_csr ='/content/drive/MyDrive/Data_viz2024_Data/crime_scene_report.csv' # ข้อมูลคดี (crime_scene_report.csv)
path_to_per ='/content/drive/MyDrive/Data_viz2024_Data/person.csv' #ข้อมูลประชากร
path_to_inter ='/content/drive/MyDrive/Data_viz2024_Data/interview.csv' # ข้อมูลการสอบสวน
path_to_income ='/content/drive/MyDrive/Data_viz2024_Data/income.csv' # ข้อมูลรายได้
path_to_member ='/content/drive/MyDrive/Data_viz2024_Data/get_fit_now_member.csv' #ข้อมูลสมาชิก ฟิตเนส Get Fit Now
path_to_get ='/content/drive/MyDrive/Data_viz2024_Data/get_fit_now_check_in.csv' #ข้อมูลการเข้าใช้ ฟิตเนส Get Fit Now
path_to_face ='/content/drive/MyDrive/Data_viz2024_Data/facebook_event_checkin.csv' #ข้อมูลการเช็คอิน Facebook
path_to_driver ='/content/drive/MyDrive/Data_viz2024_Data/drivers_license.csv' #ข้อมูลใบขับขี่
```

```python
csr_df = pd.read_csv(path_to_csr)
per_df = pd.read_csv(path_to_per)
inter_df = pd.read_csv(path_to_inter)
income_df = pd.read_csv(path_to_income)
member_df = pd.read_csv(path_to_member)
get_df = pd.read_csv(path_to_get)
face_df = pd.read_csv(path_to_face)
driver_df = pd.read_csv(path_to_driver)
```

```python
#ให้สืบหาฆาตรกร ในคดี ฆาตรกรรม (murder) ที่เกิดขึ้นในวันที่ 15 มกราคม 2018 ที่เมือง SQL City โดยเริ่มจากการตรวจสอบ ข้อมูลคดี (crime_scene_report.csv)
csr_df['date'] = csr_df['date'].astype(str)
# Use the logical AND operator between conditions
result = csr_df[(csr_df['date'].str.contains('20180115')) &
                (csr_df['type'] == 'murder') &
                (csr_df['city'] == 'SQL City')]
result
```

```python
#ภาพจากกล้องวงจรปิดแสดงให้เห็นว่ามีพยาน 2 คน พยานคนแรกอาศัยอยู่ที่บ้านหลังสุดท้ายบนถนน "Northwestern Dr"
max_number = per_df[per_df['address_street_name'] == 'Northwestern Dr']['address_number'].max()
result = per_df[(per_df['address_street_name'] == 'Northwestern Dr') & (per_df['address_number'] == max_number)]
result
```

```python
# prompt: sprit column name เเบ่งเป็นชื่อหน้าเเละหลัง
per_df[['first_name', 'last_name']] = per_df['name'].str.split(' ', n=1, expand=True)
per_df
```

```python
#ส่วนพยานคนที่สองชื่อAnnabalอาศัยอยู่ที่ไหนสักแห่งบนถนน "Franklin Ave"
per_df[(per_df['first_name'] == 'Annabel') & (per_df['address_street_name'] == 'Franklin Ave')]
```
