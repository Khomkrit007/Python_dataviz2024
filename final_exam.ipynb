```python
# quantize ข้อมูลโดยลดจำนวนข้อมูล ให้เหลือเพียง 400 x 51 โดยรวมข้อมูลที่อยู่ข้างกันและเลือกค่า max X ทำของชื่อrowด้วย จาก df_transposed

def quantize_dataframe(df, new_rows=400, new_cols=51):
  """Quantizes a DataFrame by grouping rows and selecting max values.

  Args:
    df: The input DataFrame.
    new_rows: The desired number of rows.
    new_cols: The desired number of columns.

  Returns:
    A new DataFrame with quantized data.
  """

  num_rows, num_cols = df.shape
  row_step = num_rows // new_rows
  col_step = num_cols // new_cols

  new_df = pd.DataFrame()

  for new_row_index in range(new_rows):
    start_row = new_row_index * row_step
    end_row = min((new_row_index + 1) * row_step, num_rows)

    row_data = df.iloc[start_row:end_row, :]
    row_max_values = row_data.max(axis=0)

    new_df = pd.concat([new_df, pd.DataFrame([row_max_values])], ignore_index=True)

  new_df = new_df.iloc[:, :new_cols]

  # Rename rows using original row names
  new_row_names = [df.index[min(i * row_step, num_rows - 1)] for i in range(new_rows)]
  new_df.index = new_row_names

  return new_df

# Assuming df_transposed is your DataFrame
quantized_df = quantize_dataframe(df_transposed, new_rows=400, new_cols=51)
print(quantized_df)
```

```python
# ทำ PCA ข้อมูลเพื่อให้สามารถ visualize ได้

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Assuming quantized_df is your DataFrame
X = quantized_df

# Choose the number of principal components you want to keep (e.g., 2 for visualization)
n_components = 2

# Create a PCA object
pca = PCA(n_components=n_components)

# Fit and transform the data
principalComponents = pca.fit_transform(X)

# Create a new DataFrame with the principal components
principalDf = pd.DataFrame(data=principalComponents, columns=['principal component 1', 'principal component 2'])

# Plot the principal components
plt.figure(figsize=(8, 6))
plt.scatter(principalDf['principal component 1'], principalDf['principal component 2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Quantized Data')
plt.show()

# You can also print the explained variance ratio to see how much variance each principal component explains
print(pca.explained_variance_ratio_)
```

```python
# ตอบคำถามว่า data มีทั้งหมดกี่กลุ่ม โดย plot จุดข้อมูลแต่ละกลุ่มเป็นคนละสี

# Assuming quantized_df is your DataFrame
X = quantized_df

# Choose the number of principal components you want to keep (e.g., 2 for visualization)
n_components = 2

# Create a PCA object
pca = PCA(n_components=n_components)

# Fit and transform the data
principalComponents = pca.fit_transform(X)

# Create a new DataFrame with the principal components
principalDf = pd.DataFrame(data=principalComponents, columns=['principal component 1', 'principal component 2'])

# Plot the principal components with different colors for each group (assuming you have a 'group' column)
# Replace 'group' with the actual column name representing your groups
if 'group' in quantized_df.columns:
  plt.figure(figsize=(8, 6))
  for group_name, group_data in principalDf.groupby(quantized_df['group']):
      plt.scatter(group_data['principal component 1'], group_data['principal component 2'], label=group_name)

  plt.xlabel('Principal Component 1')
  plt.ylabel('Principal Component 2')
  plt.title('PCA of Quantized Data with Group Colors')
  plt.legend()
  plt.show()

  # Print the number of unique groups
  print("Number of unique groups:", quantized_df['group'].nunique())

else:
  # If there is no 'group' column, plot without grouping
  plt.figure(figsize=(8, 6))
  plt.scatter(principalDf['principal component 1'], principalDf['principal component 2'])
  plt.xlabel('Principal Component 1')
  plt.ylabel('Principal Component 2')
  plt.title('PCA of Quantized Data')
  plt.show()
  print("No 'group' column found in the DataFrame.")

# You can also print the explained variance ratio to see how much variance each principal component explains
print(pca.explained_variance_ratio_)
```

```python
#สร้าง ตารางที่มี column บอกว่า data แต่ละจุดนั้นอยู่กลุ่มไหน

from sklearn.cluster import KMeans

# Assuming principalDf is your DataFrame with principal components
X = principalDf

# Choose the number of clusters you want to identify (e.g., 3)
n_clusters = 3

# Create a KMeans object
kmeans = KMeans(n_clusters=n_clusters, random_state=0)

# Fit the KMeans model to your data
kmeans.fit(X)

# Get the cluster labels for each data point
cluster_labels = kmeans.labels_

# Add a new column to your DataFrame indicating the cluster for each data point
principalDf['cluster'] = cluster_labels

# Print the DataFrame with the cluster information
print(principalDf)

# Plot the data points with different colors for each cluster
plt.figure(figsize=(8, 6))
for cluster_num in range(n_clusters):
  plt.scatter(
      principalDf[principalDf['cluster'] == cluster_num]['principal component 1'],
      principalDf[principalDf['cluster'] == cluster_num]['principal component 2'],
      label=f'Cluster {cluster_num + 1}'
  )

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('KMeans Clustering of PCA Data')
plt.legend()
plt.show()
```
