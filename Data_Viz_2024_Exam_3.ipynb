## 2.มีcommentsทั้งหมดเท่าใดและคอมเมนต์ไหนยอดนิยมที่สุด

```python
doc_df = pd.read_csv(path_to_doc)
doc_df
```

```python
#ให้รวมยอดไลค์ทั้งหมด
doc_df_l = doc_df.groupby('name').agg({'ยอดไลค์': 'sum', }).sum()
doc_df_l
```

```python
#คอมเมนต์ไหนยอดนิยมที่สุด
doc_df = doc_df.groupby('comments').agg({'ยอดไลค์': 'sum'}).sort_values(by='ยอดไลค์', ascending=False)
doc_df.head(1)
```
